```python
from neo4j import GraphDatabase

#uri = "bolt://localhost:7687"
#username = "neo4j"
#password = "neo4jkngbq"
uri="bolt://ec2-3-235-153-181.compute-1.amazonaws.com:7687"
username = "neo4j"
password = "somepass"
database = "neo4j"

def create_relationships_based_on_similarity(tx):
    # Find matching pairs and create relationships
    query = """
    MATCH (main1:MainNode)-[:has]->(sub1:SubNode),
          (main2:MainNode)-[:has]->(sub2:SubNode)
    WHERE main1 <> main2 AND sub1 <> sub2 AND any(key in keys(sub1) WHERE sub1[key] = sub2[key])
    CREATE (sub1)-[:RELATED_TO]->(sub2)
    """
    tx.run(query)

try:
    with GraphDatabase.driver(uri, auth=(username, password), database=database) as driver:
        with driver.session() as session:
            session.write_transaction(create_relationships_based_on_similarity)
except Exception as e:
    print(f"Error: {e}")
```

```python
from neo4j import GraphDatabase

#uri = "bolt://localhost:7687"
#username = "neo4j"
#password = "neo4jkngbq"
uri="bolt://ec2-3-235-153-181.compute-1.amazonaws.com:7687"
username = "neo4j"
password = "somepass"
database = "neo4j"

def create_relationships_between_tactics_and_software(tx):
    query = """
    MATCH (tactic:MainNode)-[:has]->(tacticSub:SubNode),
      (software:MainNode)-[:has]->(softwareSub:SubNode)
    WHERE tactic <> software AND 
          (
            toLower(softwareSub.Desc) CONTAINS toLower(tacticSub.Name) OR 
            toLower(softwareSub.Desc) CONTAINS toLower(tacticSub.Desc) OR 
            toLower(tacticSub.Desc) = toLower(softwareSub.Desc)
          )
    CREATE (tacticSub)-[:RELATED_TO]->(softwareSub)
    """
    tx.run(query)

try:
    with GraphDatabase.driver(uri, auth=(username, password), database=database) as driver:
        with driver.session() as session:
            session.write_transaction(create_relationships_between_tactics_and_software)
except Exception as e:
    print(f"Error: {e}")
```

```python
from neo4j import GraphDatabase

#uri = "bolt://localhost:7687"
#username = "neo4j"
#password = "neo4jkngbq"
uri="bolt://ec2-3-235-153-181.compute-1.amazonaws.com:7687"
username = "neo4j"
password = "somepass"
database = "neo4j"

driver = GraphDatabase.driver(uri, auth=(username, password), database=database)


def get_attack_group_and_connected_node_details(tx):
    results = []
    cursor = tx.run("MATCH (n:SubNode {filename: $filename})-[r]->(connected) RETURN n, collect(DISTINCT {node: connected, rel_type: type(r)}) AS connectedInfo", filename="attack_groups")
    
    for record in cursor:

        node = record["n"]

        connected_info = record["connectedInfo"]

        results.append({"node": node, "connectedInfo": connected_info})
    return results

def get_text_fromat_for_node(node_properties, node_text_dict):
    if(None != node_text_dict.get(node_properties.get('ID'))):

        return node_text_dict.get(node_properties.get('ID'))
    text_to_embed = ""
    if(node_properties.get('Name')):
            text_to_embed+=node_properties.get('Name')
    if(node_properties.get('Desc')):
            desc_to_add = node_properties.get('Desc')
            text_to_embed+=  "(" + desc_to_add + ")"
    node_text_dict[node_properties.get('ID')] = text_to_embed
    return text_to_embed

def get_nodes_and_text_to_embed():
    node_list = []
    with driver.session() as session:
        node_list = session.write_transaction(get_attack_group_and_connected_node_details)
        driver.close() 
    nodes_to_embed = []
    node_text_dict = {}
    
    for data in node_list:
        #this gets the node details form the result
        node = data["node"]
        #this stores both the connection nodes as its relationship with the Node as a list
        connected_info = data.get("connectedInfo")
        #get node properties like id, name, comment, description as a dictionary
        node_properties = node._properties
        
        #this will store {id:, text_to_embed:} ans will be appended to nodes_to_embed
        node_details = {}

        node_details['id'] =node_properties.get('ID')

        #text_to_embed store the textual format od the node
        text_to_embed = get_text_fromat_for_node(node_properties, node_text_dict)
        if (connected_info):
            #seperating child relation and the connected to relation
            connected_nodes_text = []
            for info in connected_info:
                connected_node = info["node"]
                rel_type = info["rel_type"]
                node_text = get_text_fromat_for_node(connected_node,node_text_dict)
                if(node_text):
                    connected_nodes_text.append(node_text)
            if(len(connected_nodes_text)>0):
                connected_text = ", ".join(connected_nodes_text) 
                text_to_embed += " using the following: " + connected_text
        
        node_details['text_to_embed'] = text_to_embed 
        nodes_to_embed.append(node_details)
        
    return nodes_to_embed
```

```python
import pandas as pd
nodes_to_embed_dicts = get_nodes_and_text_to_embed()
if(nodes_to_embed_dicts):
    df = pd.DataFrame(nodes_to_embed_dicts)
    df.to_csv("./mitre_node_texts.csv",index=False)
    print("CSV Successfully created")
```

# Load Node Embeddings to Neo4j

```python
import pandas as pd
import os

#uri = "bolt://localhost:7687"
#username = "neo4j"
#password = "neo4jkngbq"
uri="bolt://ec2-3-235-153-181.compute-1.amazonaws.com:7687"
username = "neo4j"
password = "somepass"
database = "neo4j"
driver = GraphDatabase.driver(uri, auth=(username, password))
def get_dicts_from_csv_file(file_name):
    if not os.path.isfile(file_name):
        print(f"File '{file_name}' does not exist.")
        return None
    file_path = "./"+file_name
    df = pd.read_csv(file_path)

    return df.to_dict("records")

def set_embedding_to_node(tx, node_data):
    nodeId = node_data.get('id')
    embeddings = node_data.get('embeddings')
    # Embeddings are string of floats with comma seperated values 
    # While retriving need to convert the string to list of floats
    tx.run("MATCH (node:SubNode {ID: $nodeId}) SET node.embeddings = $embeddings RETURN node",nodeId=nodeId,embeddings=embeddings)
    

def load_embeddings_from_file_to_neo4j(file_name):
    node_emb_dict_list = get_dicts_from_csv_file(file_name)
    if(None != node_emb_dict_list):
        for node_dict in node_emb_dict_list:
            if(None != node_dict.get("embeddings") and not pd.isna(node_dict.get("embeddings"))):
                try:
                    with driver.session() as session:
                        session.write_transaction(set_embedding_to_node, node_dict)

                except Exception as e:
                    print(f"Error: {e}")
        driver.close() 
    print("Loaded successsfully")
```
